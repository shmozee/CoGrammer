```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv("Cars93.csv")
# df = df.dropna()
df.head()
```

### 1. Boxplot

```python
# Boxplot for revs per mile

df_rev_per_mile = df.loc[(df['Manufacturer'] == 'Audi') 
    | (df['Manufacturer'] == 'Hyundai')
    | (df['Manufacturer'] == 'Suzuki')
    | (df['Manufacturer'] == 'Toyota')]
df_rev_per_mile
plt.figure()
sns.boxplot(x=df["Rev.per.mile"], y=df_rev_per_mile["Manufacturer"])
plt.show()


# plt.figure()
# sns.boxplot(x=df["Rev.per.mile"], y=df["Manufacturer"])
# plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Toyota

### 2. Histogram

```python
# Histogram of MPG

df_mpg = df[['MPG.city', 'MPG.highway']]
# df_mpg.head()

plt.figure()
sns.histplot(label='mpg-city', data=df_mpg, x='MPG.city', element="step", kde=True)
sns.histplot(label='mpg-highway', data=df_mpg, x='MPG.highway', element="step", kde=True)
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
More efficient on the highway.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)

df_wheelbase_turningcircle = df[['Wheelbase', 'Turn.circle']]
df_wheelbase_turningcircle.head()

plt.figure()
sns.lineplot(data=df_wheelbase_turningcircle, x='Wheelbase', y='Turn.circle')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


### 4. Barplot

```python
# Barplot for mean horsepower
# df_mean_horsepower = df.groupby('Type')['Horsepower'].mean()
# df_mean_horsepower
# plt.figure()
# sns.barplot(data=df_mean_horsepower)
# plt.show()

df_mean_horsepower = df[['Type', 'Horsepower']]
plt.figure()
sns.barplot(data=df_mean_horsepower, x='Type', y='Horsepower')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
Not nessessarily since Midsize cars have similar statistics in comparison to Large cars, so there are midsize cars with more horsepower than larger cars. But in general, smaller cars have less horsepower than all other cars.
